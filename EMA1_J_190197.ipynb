#### Problem 1 (Turning Tables)

```python
#Function Definition 

def table_html(rows):
    print("<table>")
    for n in range((rows)):
        en_plus_1 = str((n+1)) #convert formulas to string to be concactenated
        en_plus_1_t_self = str(((n+1)*(n+1)))
        if n%2 == 0:
            print('<tr class="lightyellowrow"><td>' + en_plus_1 + "</td>" + "<td>" + en_plus_1_t_self + "</td>" + "</tr>" )
        else:
            print('<tr class="whiterow"><td>' + en_plus_1 + "</td>" + "<td>" + en_plus_1_t_self + "</td>" + "</tr>" )
    print("</table>")

# Input and Input Check
while True:
    try:
        row_input = int(input("Enter number of Rows: "))
    except:
        print("Please enter a number")
        continue
    break
    
#Call the function 

table_html(row_input)
```

#### Problem 3 (Followers)

```python
def obedient(a):
    for n in range(a):
        value = 0
        m = (n+1) + (n+2) + (n+3)
        if a != m:
            value += 1
        if value > 0:
            return print("True")
        elif n == a:
            return print("False")
obedient(123)
```

#### Problem 4 (Go forth and Multiply)

```python
def multtable(size):
    for n in range(size):
        for j in range(size):
            print((j+1)*(n+1), end="    ")
        print(" ")
          
while True:
    try:
        entered_size = int(input("Size: "))
    except:
        print("Please enter a number!")
        continue
    break

multtable(entered_size)
```
