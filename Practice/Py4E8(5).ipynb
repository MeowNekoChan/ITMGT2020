```python
fname = "mbox-short.txt"
#if len(fname) < 1 : fname = "mbox-short.txt"
fh = open(fname)


for n in fh:
    print(type(n))
```

```python

fh = fh.read()
fh = str(fh)
fh = fh.split()
length_fh = len(fh)
emails = list()
count = 0
for i in range(length_fh):
    m = fh[i]
    if m == "From":
        emails.append(fh[(i+1)])
        count += 1
        
for an_email in emails:
    print(an_email)
print("There were", count, "lines in the file with From as the first word")
```
