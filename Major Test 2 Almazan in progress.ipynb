First, remove all rows containing "Landbased" and "Seabased" to leave the aggregate amounts.

```python
# Extract Country Level Rows
# Hint: mind the leading spaces in Landbased and Seabased
# Make sure you reset the index to avoid confusion


```

```python
# Write your code here to preserve the output above for comparison

df["Country"] = [i.strip() for i in df["Country"]]
not_country = ["Landbased", "Seabased"]

df_country = df[~df["Country"].isin(not_country)]
df_country = df_country.reset_index(drop = True)
df_country
```

```python
# Write code here

plt.figure(figsize = (20,12))
df_country_head = df_country_sorted.head(10)
df_country_head.reset_index(inplace = True)
df_country_head
plt.barh(y=df_country_head["Country"], width=df_country_head[2019], color='g' )
plt.gca().invert_yaxis()
plt.title("Top 10 Sources of Cash Remittances")
plt.ylabel("Country")
plt.xlabel("Remittances (in thousands)")
```

```python
plt.figure(figsize = (20,12))
plt.bar(x=df_continents_sorted["Country"], height=df_continents_sorted[2019], color='#FFA500' )
plt.title("Top 10 Continents of Cash Remittances")
plt.ylabel("Continent")
plt.xlabel("Remittances (in thousands)")
```
